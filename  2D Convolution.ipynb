
## 2D Convolution

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
im = cv2.imread('../imgs/Y2.jpg',1)
cv2.imshow("Original",im)
cv2.waitKey(0)
```

```python
# aplicando o filtro Convolução 2D
kernel = np.ones((3,3), np.float32)/9 # filtro

dst = cv2.filter2D(im,-1,kernel)

```

```python
#mostrar a imagem com o filtro

cv2.imshow("Convolução",dst)
cv2.waitKey(0)

```

```python
plt.subplot(1,2,1)
plt.title('Original')
plt.imshow(im)
plt.show()


plt.subplot(1,2,2)
plt.title('Convolução 2D')
plt.imshow(dst)
plt.show()
```

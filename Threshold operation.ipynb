```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#image 
#cv2.IMREAD_GRAYSCALE
i = cv2.imread('/media/skynet/Dados wise/git_projetos/opencv_python/imgs/gradient.jpg',1)
#cv2.imshow("GRAY",i)
#cv2.waitKey(0)

# valores para threshold


ret ,th1=  cv2.threshold(i ,127 ,255,cv2.THRESH_BINARY)
ret ,th2 = cv2.threshold(i, 127 , 255,cv2.THRESH_BINARY_INV)
ret ,th3 =  cv2.threshold(i, 127 , 255,cv2.THRESH_TRUNC)
ret , th4 = cv2.threshold(i, 127 , 255,cv2.THRESH_TOZERO)
ret , th5 = cv2.threshold(i, 127 , 255,cv2.THRESH_TOZERO_INV)
 
plt.subplot(1,1,1)
plt.title('Imagem Original')
plt.imshow(i , cmap='gray')
plt.colorbar()
plt.show()



plt.subplot(2,1,2)
plt.title('THRESH_BINARY')
plt.imshow(th1,cmap='gray')
plt.colorbar()
plt.show()


plt.subplot(2,2,1)
plt.title('THRESH_BINARY_INV')
plt.imshow(th2, cmap='gray')
plt.colorbar()
plt.show()



plt.subplot(2,2,2)
plt.title('THRESH_TRUNC')
plt.imshow(th3, cmap='gray')
plt.colorbar()
plt.show()

plt.subplot(2,3,1)
plt.title('THRESH_TOZERO')
plt.imshow(th4, cmap='gray')
plt.colorbar()
plt.show()


plt.subplot(2,3,2)
plt.title('THRESH_TOZERO_INV')
plt.imshow(th5, cmap='gray')
plt.colorbar()
plt.show()
```
